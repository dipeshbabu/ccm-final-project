```python
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
import json
```

```python
exp = []
lin = []
log = []
quad = []
quart = []

with open('../simulations/exponential_10_rounds_final_payoffs.json') as json_file:
    exp = json.load(json_file)

with open('../simulations/linear_10_rounds_final_payoffs.json') as json_file:
    lin = json.load(json_file)

with open('../simulations/logarithmic_10_rounds_final_payoffs.json') as json_file:
    log = json.load(json_file)

with open('../simulations/quadratic_10_rounds_final_payoffs.json') as json_file:
    quad = json.load(json_file)

with open('../simulations/quartic_10_rounds_final_payoffs.json') as json_file:
    quart = json.load(json_file)
```

```python
all_lists = [exp, lin, log, quad, quart]
labels = {0: 'Exponential', 1: 'Linear', 2: 'Logarithmic', 3: 'Quadratic', 4: 'Quartic'}

plt.figure(figsize=(14, 6))

for idx,data in enumerate(all_lists):
    sns.kdeplot(data, shade=True, label=labels[idx])
    #sns.kdeplot(data, shade=True, label=f'List {all_lists.index(data) + 1}')

# Add labels and title
plt.title("PDF of Payoff Values For Functions", fontsize=20)
plt.xlabel("Reward Value", fontsize=15)
plt.ylabel("Density", fontsize=15)
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Plot the boxplot with higher alpha value (transparency) and custom colors
sns.boxplot(data=all_lists)

# Add title, labels, and customize font size
plt.title("Boxplot of Payoff Values Across Functions", fontsize=20)
plt.xlabel("List", fontsize=15)
plt.ylabel("Reward Value", fontsize=15)

# Set the x-axis ticks and labels
plt.xticks(ticks=range(len(all_lists)), labels=labels.values(), fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Set up the plot with a wider figure size
plt.figure(figsize=(14, 6))

# Plot the cumulative sum for each list
for i, reward_list in enumerate(all_lists):
    cumulative_rewards = np.cumsum(reward_list)  # Compute the cumulative sum
    plt.plot(range(1, len(reward_list) + 1), cumulative_rewards, label=labels[i], alpha=0.8)

# Add title and labels
plt.title("Cumulative Progression of Reward Values Across Lists", fontsize=20)
plt.xlabel("Iteration", fontsize=15)
plt.ylabel("Cumulative Reward", fontsize=15)

# Add legend
plt.legend()

# Show grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Set up the plot with a wider figure size
plt.figure(figsize=(14, 6))

# Plot the average progression for each list
for i, reward_list in enumerate(all_lists):
    running_average = np.cumsum(reward_list) / (np.arange(1, len(reward_list) + 1))  # Compute the running average
    plt.plot(range(1, len(reward_list) + 1), running_average, label=labels[i], alpha=0.8)

# Add title and labels
plt.title("Average Progression of Reward Values Across Lists", fontsize=20)
plt.xlabel("Iteration", fontsize=15)
plt.ylabel("Average Reward", fontsize=15)

# Add legend
plt.legend()

# Show grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import glob

# Get all JSON files in the 'action_probs' folder
json_files = glob.glob('../action_probs/*.json')

# Print the list of JSON files
print(json_files)
```

```python
json_files[0].split('/')[-1].split('.')[0].split('_')[-1]
```

```python
best_actions = {}
for file in json_files:
    with open(file) as json_file:
        data = json.load(json_file)
        key = file.split('/')[-1].split('.')[0].split('_')[-1]
        best_actions[key] = data
```

```python
best_actions
```

```python
payoff_values = list( [int(num) for num in best_actions['exponential'].keys()])

plt.figure(figsize=(10, 6))

for function, payoff_dict in best_actions.items():
    proportions = list(payoff_dict.values())  # Extract hit proportions from the dictionary
    plt.plot(payoff_values, proportions, label=function)

# Adding labels and title
plt.xlabel('Payoff Value')
plt.ylabel('Proportion of Hits')
plt.title('Proportion of Hits for Each Payoff Value across Utility Functions')

# Adding legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
